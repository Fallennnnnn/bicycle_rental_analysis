## Menentukan Pertanyaan Bisnis

- Bagaimana tren peminjaman sepeda berdasarkan musim (season)?
- Apakah ada perbedaan dalam jumlah peminjaman sepeda antara hari libur (holiday) dan hari kerja (workingday)?
-  Bagaimana pengaruh cuaca (weathersit) terhadap jumlah peminjaman sepeda?

## Menyaipkan semua library yang dibuthkan

```python
# melakukan import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
df = pd.read_csv('day.csv')
```

### Assessing Data

```python
# mengecek data yang ada di dataset
df.head()
```

```python
# mengecek nilai kosong didalam data
df.isnull().sum()
```

```python
# mengecek nilai duplikat
df.duplicated()
```

### Cleaning Data

```python
# membuang data yang tidak relevan
df = df.drop(['dteday','instant','casual'], axis = 1)
```

## Exploratory Data Analysis (EDA)

### Explore

```python
# memahami data
df.describe()
```

```python
# melihat hubungan antar kolom
df.corr()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Daftar label yang akan digunakan
season_labels = ['springer', 'summer', 'fall', 'winter']

# Visualisasi jumlah peminjaman sepeda berdasarkan musim dengan x-tick yang sudah diubah
plt.figure(figsize=(8, 6))
sns.barplot(x='season', y='cnt', data=df)
plt.xlabel('Musim')
plt.ylabel('Jumlah Peminjaman')
plt.title('Jumlah Peminjaman Sepeda Berdasarkan Musim')

# Mengganti x-tick labels
plt.xticks([0, 1, 2, 3], season_labels)

plt.show()
```

### Pertanyaan 2:

```python
# Daftar label yang akan digunakan
labels = ['weekday', 'holiday']

# Visualisasi perbandingan peminjaman sepeda pada hari libur dan hari kerja
plt.figure(figsize=(8, 6))
sns.barplot(x='holiday', y='cnt', data=df)
plt.xlabel('Hari Libur')
plt.ylabel('Jumlah Peminjaman')
plt.title('Perbandingan Peminjaman Sepeda pada Hari Libur dan Hari Kerja')

# Mengganti x-tick labels
plt.xticks([0, 1], labels)

plt.show()
```

### Pertanyaan 3:

```python
# Daftar label yang akan digunakan
labels = ['Clear', 'Mist', 'Light Snow']

# Visualisasi pengaruh cuaca (weathersit) terhadap jumlah peminjaman sepeda
plt.figure(figsize=(8, 6))
sns.barplot(x='weathersit', y='cnt', data=df)
plt.xlabel('Cuaca')
plt.ylabel('Jumlah Peminjaman')
plt.title('Pengaruh Cuaca terhadap Jumlah Peminjaman Sepeda')

# Mengganti x-tick labels
plt.xticks([0, 1,2], labels)

plt.show()
```

## Conclusion

- Pada pertanyaan 1 Berdasarkan visualisasi di atas, dapat dilihat bahwa jumlah peminjam sepeda berdasarkan musim sangat beragam. Pada musim 1 yang ditandai dengan "***Springer***" terdapat jumlah peminjam paling sedikit. Sebaliknya, pada musim 3 yang ditandai dengan "***Fall***" terdapat jumlah peminjam terbanyak. Hal ini menunjukkan bahwa musim memiliki pengaruh signifikan pada jumlah peminjaman sepeda.

- Pada pertanyaan 2 Berdasarkan visualisasi di atas, dapat disimpulkan bahwa perbandingan jumlah peminjam pada saat **hari libur (holiday)** dan **hari kerja (workingday)** sangat signifikan. Hasil menunjukkan bahwa pada **hari kerja (workingday)** jumlah peminjam sepeda lebih besar dibandingkan dengan **hari libur (holiday)**. Hal ini mengindikasikan bahwa faktor hari dalam seminggu (weekday) berpengaruh signifikan terhadap jumlah peminjaman sepeda, dengan hari kerja cenderung memiliki tingkat peminjaman yang lebih tinggi.

- Pada pertanyaan 3 Berdasarkan visualisasi di atas, dapat disimpulkan bahwa perbandingan jumlah peminjam pada saat cuaca cerah ***(1: Clear, Few clouds, Partly cloudy, Partly cloudy)*** menduduki peringkat pertama dibandingkan dengan saat cuaca berkabut ***(2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist)*** dan saat ada sedikit salju ***(3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds)***. Hal ini menunjukkan bahwa cuaca cerah memiliki dampak positif terhadap jumlah peminjaman sepeda, sementara cuaca berkabut dan cuaca dengan sedikit salju cenderung memiliki tingkat peminjaman yang lebih rendah.
